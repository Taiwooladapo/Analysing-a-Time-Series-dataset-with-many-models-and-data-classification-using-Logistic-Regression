# Forecast Models

```r
train <- window(tmonthly_data, start=c(1844,1), end=c(2003,12))
test <- window(tmonthly_data, start=c(2004,1), end=c(2004,12))
```

# Arima Forecast

```r
plot(tmonthly_data,
ylab = expression("Average Temperature"),
main = "monthly data set")
```

```r
ndiffs(tmonthly_data)
```

```r
dtmonthly_data <- diff(tmonthly_data) 
plot(dtmonthly_data, xlab = "Year", ylab = "Temperature (°C)")
```

```r
adf.test(dtmonthly_data)
```

```r
Acf(dtmonthly_data)
Pacf(dtmonthly_data)
```

```r
fit <- arima(tmonthly_data, order=c(0,1,0))
fit
```

```r
fit1 <- arima(tmonthly_data, order=c(2,1,2))
fit1
```

```r
fit2 <- arima(tmonthly_data, order=c(1,1,0))
fit2
```

```r
fit3 <- arima(tmonthly_data, order=c(0,1,0))
fit3
```

```r
qqnorm(fit1$residuals)
qqline(fit1$residuals)
Box.test(fit1$residuals, type="Ljung-Box")
```

```r
fit1 <- auto.arima(train) 
fit1
```

```r
fam <- forecast(fit1, 12)
fam
```

```r
plot(fam)
```

```r
plot(fam, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fam, test)
```

# ETS Forecast

```r
em <- ets(train, model='ZZZ')
em
```

```r
fem <- forecast(em, h=12) 
fem
```

```r
plot(fem)
```

```r
plot(fem, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fem, test)
```

# Simple Time Series

# Mean Forecast

```r
mean(train)
```

```r
fcast.mean <- meanf(train,h=12)
fcast.mean
```

```r
accuracy(fcast.mean, test)
```

```r
plot(fcast.mean)
```

```r
plot(fcast.mean, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

# Naive or Random Walk Forecast

```r
train[length(train)]
```

```r
fcast.naive <- naive(train,h=12)
fcast.naive
```

```r
accuracy(fcast.naive, test)
```

```r
plot(fcast.naive)
```

```r
plot(fcast.naive, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

# Direct Comparison

```r
par(mfrow=c(2,2))

plot(fam, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)

plot(fem, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)

plot(fcast.mean, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)

plot(fcast.naive, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fam, test) 
accuracy(fem, test)
accuracy(fcast.mean, test)
accuracy(fcast.naive, test)
```
