# Forecast Models

```r
train <- window(tyearly_data, start=c(1844,1), end=c(2003,1))
test <- window(tyearly_data, start=c(2004,1), end=c(2004,1))
```

# Arima Forecast

```r
plot(tyearly_data,
ylab = expression("Average Temperature"),
main = "yearly data set")
```

```r
ndiffs(tyearly_data)
```

```r
dtyearly_data <- diff(tyearly_data, lag=12) 
plot(dtyearly_data, xlab = "Year", ylab = "Temperature (°C)")
```

```r
adf.test(dtyearly_data)
```

```r
Acf(dtyearly_data)
Pacf(dtyearly_data)
```

```r
fit <- arima(tyearly_data, order=c(0,1,0))
fit
```

```r
fit1 <- arima(tyearly_data, order=c(2,1,2))
fit1
```

```r
fit2 <- arima(tyearly_data, order=c(1,1,0))
fit2
```

```r
fit3 <- arima(tyearly_data, order=c(0,1,0))
fit3
```

```r
qqnorm(fit1$residuals)
qqline(fit1$residuals)
Box.test(fit1$residuals, type="Ljung-Box")
```

```r
fit1 <- auto.arima(train) 
fit1
```

```r
fam <- forecast(fit1, 1)
fam
```

```r
plot(fam)
```

```r
plot(fam, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fam, test)
```

# ETS Forecast

```r
em <- ets(train, model='ZZZ')
em
```

```r
fem <- forecast(em, h=1) 
fem
```

```r
plot(fem)
```

```r
plot(fem, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fem, test)
```

# Simple Time Series

# Mean Forecast

```r
mean(train)
```

```r
fcast.mean <- meanf(train,h=1)
fcast.mean
```

```r
plot(fcast.mean)
```

```r
plot(fcast.mean, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fcast.mean, test)
```

# Naive or Random walk Forecast

```r
train[length(train)]
```

```r
fcast.naive <- naive(train,h=1)
fcast.naive
```

```r
plot(fcast.naive)
```

```r
plot(fcast.naive, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
accuracy(fcast.naive, test)
```

# Direct Comparison

```r
par(mfrow=c(2,2))

plot(fam, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)

plot(fem, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)

plot(fcast.mean, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)

plot(fcast.naive, xlim=c(2003,2004.95), ylim=c(0,20))
x <- as.vector(time(test))
y <- test
lines(x, y, col='red', lwd=3)
```

```r
par(mfrow=c(2,2))

plot(fam)

plot(fem)

plot(fcast.mean)

plot(fcast.naive)
```

```r
accuracy(fam, test) 
accuracy(fem, test)
accuracy(fcast.mean, test)
accuracy(fcast.naive, test)
```
